```python
import pandas as pd
import numpy as np
import plotly.express as px
import datetime
from scipy import stats
```

```python
df = pd.read_pickle('dww_scraper_export')

df2 = df['Concentration level'].str.split(" ", n = 1, expand = True)
df2.rename(columns = {0:'Concentration Measure', 1:'Concentration Unit'}, inplace=True)
df3 = df.join(df2)

#remove blank spaces from data
column = list(df3)
for i in column:
    df3[i] = df3[i].str.strip()
df3
```

```python
#Change measurement data type to numeric
df3['Concentration Measure'] = pd.to_numeric(df3['Concentration Measure'])
df3.dtypes
```

```python
#Change date data to datetime format
df3['Collection Date'] = pd.to_datetime(df3['Collection Date'], format= '%m-%d-%Y')
```

```python
df3.to_csv('DWW_Scraper_Data_Export_2-11-2020.csv')
```
